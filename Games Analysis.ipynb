```python
game = game[game['Year'].notnull()]
```

```python
round(game.isnull().sum().sort_values(ascending=False)/len(game)*100,2)
```

```python
X=game.iloc[:,0:4]
X
```

```python
y=game.iloc[:,-1]
y
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
LR.score(X_test,y_test)*100
```

```python
y_pred = LR.predict(X_test)
y_pred
```

```python
def predict_Global_Sales(Year,NA_Sales,EU_Sales,JP_Sales):
    
    x=np.zeros(len(X.columns))
    x[0] = Year
    x[1] = NA_Sales
    x[2] = EU_Sales
    x[3] = JP_Sales
    
    return LR.predict([x])[0]
    
```
